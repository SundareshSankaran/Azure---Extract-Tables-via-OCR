# Start Analyzing

## Local File

#### image

```python
import requests

service_url = 'http://localhost:5000/'

# Define the URL and file path
url = f"{service_url}formrecognizer/documentModels/prebuilt-layout:analyze?api-version=2022-08-31"
file_path = "test-letter.jpg"

# Define headers
headers = {
    'accept': '*/*',
    'Content-Type': 'application/octet-stream',
}

# Open the file and send the POST request
with open(file_path, 'rb') as file:
    response = requests.post(url, headers=headers, data=file)

# Print the response
print(response.status_code)
print(response.headers.get('Operation-Location'))
```

#### pdf

```python
import requests

service_url = 'http://localhost:5000/'

# Define the URL and file path
url = f"{service_url}formrecognizer/documentModels/prebuilt-layout:analyze?api-version=2022-08-31"
file_path = "Krankenhaus_Wikipedia.pdf"

# Define headers
headers = {
    'accept': '*/*',
    'Content-Type': 'application/octet-stream',
}

# Open the file and send the POST request
with open(file_path, 'rb') as file:
    response = requests.post(url, headers=headers, files={'file': file})

# Print the response
print(response.status_code)
print(response.headers.get('Operation-Location'))
```

## Online File

```python
service_url = 'http://localhost:5000/'

# Define the URL and JSON payload
url = f"{service_url}formrecognizer/documentModels/prebuilt-layout:analyze?api-version=2022-08-31"

payload = {
    "urlSource": "https://raw.githubusercontent.com/Azure-Samples/cognitive-services-REST-api-samples/master/curl/form-recognizer/rest-api/layout.png"
}

# Define headers
headers = {
    'accept': '*/*',
    'Content-Type': 'application/json',
}

# Send the POST request
response = requests.post(url, json=payload, headers=headers)

# Print the response
print(response.status_code)
print(response.headers.get('Operation-Location'))
```

# Get Result

```python
import requests, json

# Define the URL
url = response.headers.get('Operation-Location')

# Define headers
headers = {
    'accept': '*/*',
}

# Send the GET request
response_get = requests.get(url, headers=headers)

# Print the response
#print(response_get.status_code)
#print(response_get.text)

status = json.loads(response_get.text)['status']

if(status=='running'):
    print("stil processing")
else:
    print("done")
```
