```python
import requests
import os
import dotenv
import json
import base64


dotenv.load_dotenv()
```

```python
KEY = os.getenv('OCR_KEY')
ENDPOINT = os.getenv('OCR_ENDPOINT')
VERSION = '2023-10-31-preview'
MODEL_ID = 'prebuilt-layout'
```

# Simple Example

```python

DOCUMENT_URL = 'https://raw.githubusercontent.com/Azure-Samples/cognitive-services-REST-api-samples/master/curl/form-recognizer/rest-api/layout.png'  # URL of the document to analyze

url = f"{ENDPOINT}documentintelligence/documentModels/{MODEL_ID}:analyze?_overload=analyzeDocument&api-version={VERSION}"
headers = {
    "Content-Type": "application/json",
    "Ocp-Apim-Subscription-Key": KEY
}
data = {
    "urlSource": DOCUMENT_URL
}

response = requests.post(url, headers=headers, json=data)

print(response.status_code)
print(response.text)
```

```python
response_url = response.headers.get('Operation-Location')  # This is the URL you got from the POST response
headers = {
    "Ocp-Apim-Subscription-Key": KEY
}

response = requests.get(response_url, headers=headers)

# Assuming the response content is JSON, which is common for APIs
response_json = response.json()

# Pretty print the JSON response+
print(json.dumps(response_json, indent=4))
```

```python
# 'suceeded' if finished, 'running' or 'notStarted' if not
response_json.get('status')
```

```python
import base64

# Replace 'path/to/your/image.jpg' with the actual path to your JPEG file
file_path = 'data/german-handwriting-sample.jpeg'

# Open the file in binary read mode
with open(file_path, 'rb') as image_file:
    # Read the file content
    image_data = image_file.read()
    # Encode file content to base64
    base64_encoded_data = base64.b64encode(image_data)

    # If you need the base64 string in a text format for JSON or other uses
    base64_string = base64_encoded_data.decode('utf-8')

print(base64_string)
```

```python

# Construct the full URL with query parameters
url = f"{ENDPOINT}documentintelligence/documentModels/{MODEL_ID}:analyze?_overload=analyzeDocument&api-version={VERSION}"

# Request headers
headers = {
    "Ocp-Apim-Subscription-Key": KEY,
    "Content-Type": "application/json"
}

# Request body
data = {
    "base64Source": base64_string
}

response = requests.post(url, headers=headers, json=data)

print(response.status_code)
print(response.text)
```
