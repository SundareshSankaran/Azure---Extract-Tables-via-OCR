```python
import requests

# Replace these variables with your actual values
ocr_key = os.getenv('OCR_KEY')
ocr_endpoint = os.getenv('OCR_ENDPOINT')
VERSION = '2023-10-31-preview'
MODEL_ID = 'prebuilt-layout'  # Your specific model ID
DOCUMENT_URL = 'https://raw.githubusercontent.com/Azure-Samples/cognitive-services-REST-api-samples/master/curl/form-recognizer/rest-api/layout.png'  # URL of the document to analyze

url = f"{ocr_endpoint}documentintelligence/documentModels/{MODEL_ID}:analyze?_overload=analyzeDocument&api-version=2023-10-31-preview"
headers = {
    "Content-Type": "application/json",
    "Ocp-Apim-Subscription-Key": ocr_key
}
data = {
    "urlSource": DOCUMENT_URL
}

response = requests.post(url, headers=headers, json=data)

print(response.status_code)
print(response.text)
```
